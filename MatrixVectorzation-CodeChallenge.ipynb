```python
R_df = df_train.pivot(index = 'user_id', columns ='challenge', values = 'challenge_sequence').fillna(0)
R_df.head()
```

```python
R = R_df.as_matrix()
user_ratings_mean = np.mean(R, axis = 1)
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
```

```python
R_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 50)
```

```python
sigma = np.diag(sigma)
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = R_df.columns)
```

```python
def recommend_challenge(predictions_df, userID, challenge_df, original_ratings_df, num_recommendations=3):
    
    # Get and sort the user's predictions
    user_row_number = userID # UserID starts at 1, not 0
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data and merge in the movie information.
    user_data = original_ratings_df[original_ratings_df.user_id == (userID)]
    user_full = (user_data.merge(challenge_df, how = 'left', left_on = 'challenge', right_on = 'challenge').
                     sort_values(['challenge_sequence'], ascending=True)
                 )

    #print ('User {0} has already rated {1} challenges.',format(userID, user_full.shape[0]))
    #print ('Recommending the highest {0} predicted ratings challenges not already rated.',format(num_recommendations))
    
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (challenge_df[~challenge_df['challenge'].isin(user_full['challenge'])].
         merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
               left_on = 'challenge',
               right_on = 'challenge').
         rename(columns = {user_row_number: 'Predictions'}).
         sort_values('Predictions', ascending = True).
                       iloc[:num_recommendations, :-1]
                      )

    return user_full, recommendations['challenge']

already_rated, predictions = recommend_challenge(preds_df, 4577, df_challenge, df_train, 10)
```
